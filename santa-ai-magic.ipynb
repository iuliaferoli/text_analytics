### APIs setup

```
import requests,uuid, pprint

#Text analytics
analytics_key = YOURKEYHERE
assert analytics_key
text_analytics_base_url = "https://westeurope.api.cognitive.microsoft.com/text/analytics/v2.0/"

language_api_url = text_analytics_base_url + "languages"

#Sentiment
sentiment_api_url = text_analytics_base_url + "sentiment"

#Key phrases
key_phrase_api_url = text_analytics_base_url + "keyPhrases"


#Translation
translate_key = YOURKEYHERE
endpoint_translate = "https://api.cognitive.microsofttranslator.com"
location = "westeurope"
path = '/translate'

constructed_url = endpoint_translate + path
```

### Document to use as example if you're not connecting to the data lake with previous steps

```
text = """Hello Santa, 

This year is very sad for us because we need to stay at home all the time. I’m crying a lot in my room. :-( 
Maybe you can make us happy with a nice present? I would like a car and a big fluffy teddy bear. 

Thank you, 

Little Jimmy
If you don’t have our address, it’s: Evert van de Beekstraat 354, 1118 CZ Schiphol"""
```

```
documents = { 'documents': [
    { 'id': '1', 'text': text}]}
```

### Language detection

```
headers_text   = {"Ocp-Apim-Subscription-Key": analytics_key}
response  = requests.post(language_api_url, headers=headers_text, json=documents)
languages = response.json()

wishes["detected_language"] = languages["documents"][0]['detectedLanguages'][0]['name']
wishes["detected_language"]
```

### Translation to dutch

```
body = [{
      'text': text}]
headers_translate = {
    'Ocp-Apim-Subscription-Key': translate_key,
    'Ocp-Apim-Subscription-Region': location,
    'Content-type': 'application/json',
    'X-ClientTraceId': str(uuid.uuid4())
}
params = {
    'api-version': '3.0',
    'to': ['nl']
}

request = requests.post(constructed_url, params=params, headers=headers_translate, json=body)
response = request.json()

wishes["translated_data"] = response[0]['translations'][0]["text"]
wishes["translated_data"] 
```

### Sentiment analysis

```
headers_sentiment   = {"Ocp-Apim-Subscription-Key": analytics_key}
response  = requests.post(sentiment_api_url, headers=headers_sentiment, json=documents)
sentiments = response.json()

wishes["emotion_level"] = sentiments["documents"][0]['score']
wishes["emotion_level"]
```

### Key phrase

```
headers_key   = {"Ocp-Apim-Subscription-Key": analytics_key}
response  = requests.post(key_phrase_api_url, headers=headers_key, json=documents)
key_phrases = response.json()
print(key_phrases)
```
